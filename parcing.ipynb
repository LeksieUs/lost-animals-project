```python
import requests
from bs4 import BeautifulSoup
from collections import defaultdict 
import time
from google.colab import files
from urllib.parse import urlparse, urljoin
import pandas as pd
```

```python
headers = {
    'Access-Control-Allow-Origin': '*',
    'Access-Control-Allow-Methods': 'GET',
    'Access-Control-Allow-Headers': 'Content-Type',
    'Access-Control-Max-Age': '3600',
    'User-Agent': 'Mozilla/5.0 (X11; Ubuntu; Linux x86_64; rv:52.0) Gecko/20100101 Firefox/52.0'
}

url = 'https://books.toscrape.com/index.html'

```

```python
s = """Travel
Mystery
Historical Fiction
Sequential Art
Classics
Philosophy
Romance
Womens Fiction
Fiction
Childrens
Religion
Nonfiction
Music
Default
Science Fiction
Sports and Games
Add a comment
Fantasy
New Adult
Young Adult
Science
Poetry
Paranormal
Art
Psychology
Autobiography
Parenting
Adult Fiction
Humor
Horror
History
Food and Drink
Christian Fiction
Business
Biography
Thriller
Contemporary
Spirituality
Academic
Self Help
Historical
Christian
Suspense
Short Stories
Novels
Health
Politics
Cultural
Erotica
Crime"""
s = s.split("\n")
lst_of_genres = []
for i in range(len(s)):
    x = s[i].split()
    lst_of_genres.append(x)
print(lst_of_genres)
```

```python
books = []
counter = 1
for i in range(50):
    
    url = 'https://books.toscrape.com/catalogue/category/books_1/page-' + str(counter) + ".html"
    req = requests.get(url, headers)
    soup = BeautifulSoup(req.content, 'html.parser')
    t = soup.find_all('a', href=True)
    for i in t:
        x = i.text
        x = x.split()
        if x in lst_of_genres:
            continue 
        elif x == "\n" or x == []: # бывают лишние пустые строки
            continue 
        else:
            books.append(x)
    counter += 1
print(books)
```

```python
counter = 1
resolvedURLs = []
for i in range(50):
    
    url = 'https://books.toscrape.com/catalogue/category/books_1/page-' + str(counter) + ".html"
    getURL = requests.get(url, headers={"User-Agent":"Mozilla/5.0"})
    soup = BeautifulSoup(getURL.text, 'html.parser')
    images = soup.find_all('img')
    
    for image in images:
        src = image.get('src')
        resolvedURLs.append(requests.compat.urljoin(url, src))
    
            
    counter += 1
print(resolvedURLs)


```

```python
prices = []
counter = 1
for i in range(50):
    
    url = 'https://books.toscrape.com/catalogue/category/books_1/page-' + str(counter) + ".html"
    req = requests.get(url, headers)
    soup = BeautifulSoup(req.content, 'html.parser')
    t = soup.find_all('p')
    for i in t:
        x = i.text
        x = x.split()
        if x == "\n" or x == []: # бывают лишние пустые строки
            continue 
        elif x == ['In', 'stock']:
            continue
        else:
            prices.append(x)
    counter += 1
print(prices)



```

```python

stars = []
counter = 1
for i in range(50):
    
    url = 'https://books.toscrape.com/catalogue/category/books_1/page-' + str(counter) + ".html"
    req = requests.get(url, headers)
    soup = BeautifulSoup(req.content, 'html.parser')
    t = soup.find_all('p')

    for i in t:
        i = str(i)
        if i[23] == "h":
            stars.append(3)
        elif i[22] == "O":
            stars.append(1)
        elif i[23] == "w":
            stars.append(2)
        elif i[24] == "u":
            stars.append(4)
        elif i[23] == "i":
            stars.append(5)



    
    counter += 1
print(stars)
print(len(stars))
```

```python
links = []

counter = 1
for i in range(50):
    
    url = 'https://books.toscrape.com/catalogue/category/books_1/page-' + str(counter) + ".html"
    req = requests.get(url, headers)
    soup = BeautifulSoup(req.content, 'html.parser')
    t = soup.find_all('a', href = True)
    for a_tag in t:
        href = a_tag.attrs.get("href")
        href = urljoin(url, href)
        parsed_href = urlparse(href)
        href = parsed_href.scheme + "://" + parsed_href.netloc + parsed_href.path
        if href not in links:
            s = str(href)
            if "https://books.toscrape.com/catalogue/category/" not in s:
                links.append(href)

    
    counter += 1
links = links[1::]
print(len(links))
print(links)
```

```python

books_full = []

for url in links[0:100]:
    req = requests.get(url, headers)
    soup = BeautifulSoup(req.content, 'html.parser')

    t = soup.find_all('h1')
    for i in t:
        x = i.text
        books_full.append(x)
```

```python
available = []
for url in links[0:100]:
    req = requests.get(url, headers)
    soup = BeautifulSoup(req.content, 'html.parser')

    t = soup.find_all('p')
    for i in t:
        x = i.text
        x = x.split()
        if x == "\n" or x == []: # бывают лишние пустые строки
            continue 
        if len(x) > 2:
            if x[0] == "In":
                if str(x[2])[1::].isdigit():
                    available.append(str(x[2])[1::])
print(available)
print(len(available))
```

```python
table_all = []
for url in links[0:100]:
    req = requests.get(url, headers)
    soup = BeautifulSoup(req.content, 'html.parser')

    table = {}
    t1 = soup.find_all('th')
    t2 = soup.find_all('td')
    for i in range(len(t1)):
        k = t1[i].text
        v = t2[i].text
        table[k] = v
    table_all.append(table)
```

```python
description = []
for url in links[0:100]:
    req = requests.get(url, headers)
    soup = BeautifulSoup(req.content, 'html.parser')

    t = soup.find_all('p')
    t = t[-1]
    for i in t:
        description.append(i)
```

```python

gener = []
for url in links[0:100]:   
    req = requests.get(url, headers)
    soup = BeautifulSoup(req.content, 'html.parser')
    
    t = soup.find_all('a', href = True)
    for i in t:
        x = i.text
        if x.split() == "\n" or x.split() == []: # бывают лишние пустые строки
            continue 

        if x != "Books to Scrape" and x != "Home" and x != "Books" and x.split() not in books:
            gener.append(x)

print(gener)
print(len(gener))
```

```python
data = []
print(len(books_full), len(gener), len(available), len(stars), len(links), len(prices))
for i in range(99):
    x = [books_full[i], gener[i], available[i], stars[i], "img" + str(i + 1), links[i], str(prices[i][0])]
    j = table_all[i]
    
    for k in j:

        x.append(j[k])
    data.append(x)

```

```python
columns = ["name", "gener", "available", "stars", "picture", "link", "price", "UPC", "type", "price_excl_tax", "price_incl_tax", "tax", "availability", "reviews"]
df = pd.DataFrame(data, columns=columns)
df.head()
```
